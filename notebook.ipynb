```python
columns_to_delete = ['Unnamed: 0', 'borough', 'Borough', 'BldgClass', 'apartment_number', 'building_class_category', 'ZipCode',
                    'YearBuilt', 'BoroCode', 'Address', 'easement', 'FIRM07_FLA', 'PFIRM15_FL', 'Ext', 'HistDist', 'Landmark',
                    'LtdHeight', 'Overlay1', 'Overlay2', 'OwnerType', 'SPDist1', 'SPDist2', 'SPDist3', 'ZMCode', 'ZoneDist2',
                    'ZoneDist3', 'ZoneDist4', 'EDesigNum', 'CT2010', 'CB2010', 'AreaSource', 'UnitsRes', 'UnitsTotal', 'Tract2010',
                    'APPDate']
df.drop(columns_to_delete, inplace=True, axis=1)
```

Удалим строки с ценой продажи 0, так как скорее всего, нулевая цена продажи не связана с продажей (к тому же скорее всего sale_price окажется таргетной переменной). Также заменим NaN/0 во многих строках на медиану/моду.

```python
df=df[df['sale_price'] != 0]
df['gross_sqft'] = df['gross_sqft'].replace(0.0, df['gross_sqft'].median())
df['land_sqft'] = df['land_sqft'].replace(0.0, df['land_sqft'].median())
df['NumBldgs']= df['NumBldgs'].fillna(df['NumBldgs'].median())
df['NumFloors']= df['NumFloors'].fillna(df['NumFloors'].median())
df['ProxCode']= df['ProxCode'].fillna(df['ProxCode'].mode()[0])
df['LotType']= df['LotType'].fillna(df['LotType'].mode()[0])
df['BsmtCode']= df['BsmtCode'].fillna(df['BsmtCode'].mode()[0])
df['LandUse']= df['LandUse'].fillna(df['LandUse'].mode()[0])
df['AssessLand']= df['AssessLand'].fillna(df['AssessLand'].median())
df['AssessTot']= df['AssessTot'].fillna(df['AssessTot'].median())
df['ExemptLand']= df['ExemptLand'].fillna(df['ExemptLand'].median())
df['ExemptTot']= df['ExemptTot'].fillna(df['ExemptTot'].median())
df['BuiltFAR']= df['BuiltFAR'].fillna(df['BuiltFAR'].median())
df['ResidFAR']= df['ResidFAR'].fillna(df['ResidFAR'].median())
df['CommFAR']= df['CommFAR'].fillna(df['CommFAR'].median())
df['FacilFAR']= df['FacilFAR'].fillna(df['FacilFAR'].mean())
df['OwnerName']= df['OwnerName'].fillna(value=0)
df['IrrLotCode']= df['IrrLotCode'].fillna(value=0)
df['SplitZone']= df['SplitZone'].fillna(value=0)
df['XCoord']= df['XCoord'].fillna(df['XCoord'].mode()[0])
df['YCoord']= df['YCoord'].fillna(df['YCoord'].mode()[0])
df['XCoord']= df['XCoord'].replace(0.0,df['XCoord'].mode()[0] )
df['YCoord']= df['YCoord'].replace(0.0,df['YCoord'].mode()[0] )
df['ComArea']= df['ComArea'].fillna(df['ComArea'].median())
df['ResArea']= df['ResArea'].fillna(df['ResArea'].median())
df['OfficeArea']= df['OfficeArea'].fillna(df['OfficeArea'].median())
df['RetailArea']= df['RetailArea'].fillna(df['RetailArea'].median())
df['GarageArea']= df['GarageArea'].fillna(df['GarageArea'].median())
df['OtherArea']= df['OtherArea'].fillna(df['OtherArea'].median())
df['StrgeArea']= df['StrgeArea'].fillna(df['StrgeArea'].median())
df['FactryArea']= df['FactryArea'].fillna(df['FactryArea'].median())
df['LotFront']= df['LotFront'].fillna(df['LotFront'].median())
df['LotDepth']= df['LotDepth'].fillna(df['LotDepth'].median())
df['BldgFront']= df['BldgFront'].fillna(df['BldgFront'].median())
df['BldgDepth']= df['BldgDepth'].fillna(df['BldgDepth'].median())
df['HealthArea']= df['HealthArea'].fillna(df['HealthArea'].median())
df['YearAlter1']= df['YearAlter1'].fillna(df['YearAlter1'].mode()[0])
df['YearAlter2']= df['YearAlter2'].fillna(df['YearAlter2'].mode()[0])
df['SchoolDist']= df['SchoolDist'].fillna(df['SchoolDist'].median())
df['Council']= df['Council'].fillna(df['Council'].median())
df['PolicePrct']= df['PolicePrct'].fillna(df['PolicePrct'].median())
df['HealthCent']= df['HealthCent'].fillna(df['HealthCent'].median())
df['SanitDistr']= df['SanitDistr'].fillna(df['SanitDistr'].median())
df['Easements']= df['Easements'].fillna(df['Easements'].median())
df['LotArea']= df['LotArea'].fillna(df['LotArea'].median())
df['BldgArea']= df['BldgArea'].fillna(df['BldgArea'].median())
df['BBL']= df['BBL'].fillna(df['BBL'].median())
df['CondoNo']= df['CondoNo'].fillna(df['CondoNo'].median())
df['TaxMap']= df['TaxMap'].fillna(df['TaxMap'].median())
df['APPBBL']= df['APPBBL'].fillna(df['APPBBL'].median())
df['PLUTOMapID']= df['PLUTOMapID'].fillna(df['PLUTOMapID'].median())
df['MAPPLUTO_F']= df['MAPPLUTO_F'].fillna(df['MAPPLUTO_F'].median())
df['SHAPE_Leng']= df['SHAPE_Leng'].fillna(df['SHAPE_Leng'].median())
df['SHAPE_Area']= df['SHAPE_Area'].fillna(df['SHAPE_Area'].median())
```
